```python
result_list=[]

def get_recipe_list2(food):
    base_url='https://www.10000recipe.com/recipe/list.html'
    
    food=input('재료를 입력해주세요:').split(',')
    
    params={
        'q':food
    }
    
    resp=requests.get(base_url,params=params)
    soup=BeautifulSoup(resp.text,'lxml')
    
    recipe_list_tags=soup.select('ul.common_sp_list_ul>li')
    
    for recipe in recipe_list_tags:
        recipe={'title':recipe_list_tags.find_all('div').get('common_sp_caption_tit'),
               'url':recipe_list_tags.find('a').get('href')}
        
        result_list.append(recipe)
    
    
    return result_list
```

```python
get_recipe_list2('계란')
```
