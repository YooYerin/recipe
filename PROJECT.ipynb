## 추천순 기준 첫번째 페이지 title, url 가져오기

```python
a = input("재료를 입력하세요: ")

value = a.split(',')

q = ''
for val in value:
    q = q + val + ' '

q = q[:-1]
q
```

```python
url = 'http://www.10000recipe.com/recipe/list.html'

menu_list = []
for i in range(1):
    params = {
        'q': q,
        'page': i+1 
    }
    resp = requests.get(url, params=params)
    soup = BeautifulSoup(resp.text)
    ul_tag = soup.find_all('ul', class_='common_sp_list_ul ea4')
    
    for ul in ul_tag:        
        div_tags = ul.select('li div.common_sp_caption_tit.line2')  # title 추출 위함
        a_tags = ul.select('li>div>a')  # url 추출 위함

        for div, a in zip(div_tags, a_tags):
            menu_list.append({
                'title': div.text,
                'url': a.get('href')
            })
    

print(len(menu_list))
menu_list        
```

## detail 가져오기

```python
base_url = 'http://www.10000recipe.com/recipe/list.html'

menu_list = []
for i in range(1):
    params = {
        'q': q,
        'page': i+1 
    }
    resp = requests.get(base_url, params=params)
    soup = BeautifulSoup(resp.text)
    ul_tag = soup.find_all('ul', class_='common_sp_list_ul ea4')
    
    for ul in ul_tag:        
        div_tags = ul.select('li div.common_sp_caption_tit.line2')  # title 추출 위함
        a_tags = ul.select('li>div>a')  # url 추출 위함

        for div, a in zip(div_tags, a_tags):
            menu_list.append({
                'title': div.text,
                'url': a.get('href')
            })
print(len(menu_list))
menu_list        
```

## 시간, 난이도

```python
time_list = []
level_list = []

for url in url_list:
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)
    
    div_tags = soup.find_all('div', class_='view2_summary_info')
    
    for div in div_tags:
        time_tag = div.select('span.view2_summary_info2')
        level_tag = div.select('span.view2_summary_info3')
        
        for time in time_tag:
            time_list.append(time.text)
        for level in level_tag:
            level_list.append(level.text)
        
print(time_list)
print(level_list)
```

```python
len(time_list)
```

## 재료 리스트

```python
ingredient_list = []
blank_list = []
for url in url_list:
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)
    
    div_tag = soup.find_all('div', class_='ready_ingre3')  
    if not div_tag:
        ingredient_list.append([])
        continue
        
    
    temp_list = []
    for div in div_tag:
        li_tag = div.select('ul > a > li')
        for li in li_tag:
            name = li.text.strip('\n')
            source = name.split('\n')
            if len(source) == 1:
                source = [source[0].strip(), None]
            else:
                source = [source[0].strip(), source[1].strip()]
            
            
            temp_list.append(source)
            if li == li_tag[len(li_tag)-1]:
                ingredient_list.append(temp_list)
                temp_list = []
        
print(len(ingredient_list))
ingredient_list
```

```python
len(ingredient_list)
```

#### 재료만 따로 저장

```python
ingredients = []

for i in ingredient_list:
    temp = []
    for j in i:
        food = j[0]
        temp.append(food)
    ingredients.append(temp)
ingredients
```
